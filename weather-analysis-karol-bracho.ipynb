# Question 1: What is the average temperature for each month?
## Steps:
  1. Convert the `Formatted Date`to `datetime`
  2. Extract the month from the `Formatted Date`
  3. Group by month to calculate the average temperature.

```python
# 1. Convert to datetime
df["Formatted Date"] = pd.to_datetime(df["Formatted Date"], utc=True)
df.head()
```

```python
# 2. Extract the month and year from the "Formatted Date"
df["Year"] = df["Formatted Date"].dt.year
df["Month"] = df["Formatted Date"].dt.month
df.head()
```

```python
# 3. Calculate the average temperature for each month
monthly_avg_temp = df.groupby("Month")["Temperature (C)"].mean().reset_index()
print("Average Temperature for each month:")
print(monthly_avg_temp)
```

# Question 2: How does humidity correlate with temperature?
## Steps:
  1. We are taking advantage of steps 1 and 2 from previous question (conerting and extracting `Formatted Date`).
  2. Calculate the correlation between humidity and temperature

```python
# 3. Calculate the correlation between humidity and temperature
humidity_temp_correlation = df["Humidity"].corr(df["Temperature (C)"])
print("Correlation between Humidity and Temperature:")
print(humidity_temp_correlation)
```
